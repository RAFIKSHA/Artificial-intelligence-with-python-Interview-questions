```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression,LinearRegression,Lasso,Ridge,LogisticRegression,RidgeClassifier
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,train_test_split

from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
data=pd.read_csv('HousingData.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['CRIM']=data['CRIM'].fillna(np.mean(data['CRIM']))           
data['ZN']=data['ZN'].fillna(np.mean(data['ZN']))     
data['INDUS']=data['INDUS'].fillna(np.mean(data['INDUS']))
data['CHAS']=data['CHAS'].fillna(np.mean(data['CHAS']))
data['AGE']=data['AGE'].fillna(np.mean(data['AGE']))
data['LSTAT']=data['LSTAT'].fillna(np.mean(data['LSTAT']))
```

```python
data['CRIM'].isna().sum()
data['ZN'].isna().sum()
data['INDUS'].isna().sum()
data['CHAS'].isna().sum()
data['AGE'].isna().sum()
data['LSTAT'].isna().sum()
```

# Train_Test_Split :

```python
y=data['MEDV']                  #Dependent Data
x=data.drop('MEDV',axis=1)      #Independent Data

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=12,test_size=0.25)
```

```python
y,x
```

```python
x_train
```

# Linear Regression Model

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
#Testing Data Evaluation
y_pred=lr_model.predict(x_test)
print(y_pred)

mse=mean_squared_error(y_test,y_pred)
print('MSE = ',mse)

rmse=np.sqrt(mse)
print('RMSE = ',rmse)

mae=mean_absolute_error(y_test,y_pred)
print('MAE = ',mae)

score=r2_score(y_test,y_pred)
print('Score = ',score*100,'%')
```

```python
#Training Data Evaluation
y_train_pred=lr_model.predict(x_train)
print(y_train_pred)

mse=mean_squared_error(y_train,y_train_pred)
print('MSE = ',mse)

rmse=np.sqrt(mse)
print('RMSE = ',rmse)

mae=mean_absolute_error(y_train,y_train_pred)
print('MAE = ',mae)

score=r2_score(y_train,y_train_pred)
print('Score = ',score*100,'%')
```

# Lasso Regression 

```python
lasso_model=Lasso(alpha=1.0)
lasso_model.fit(x_train,y_train)
```

```python
#Testing Data Evaluation
y_pred=lasso_model.predict(x_test)
print(y_pred)

mse=mean_squared_error(y_test,y_pred)
print('MSE = ',mse)

rmse=np.sqrt(mse)
print('RMSE = ',rmse)

mae=mean_absolute_error(y_test,y_pred)
print('MAE = ',mae)

score=r2_score(y_test,y_pred)
print('Score = ',score*100,'%')
```

```python
#Training Data Evaluation
y_train_pred=lasso_model.predict(x_train)
print(y_train_pred)

mse=mean_squared_error(y_train,y_train_pred)
print('MSE = ',mse)

rmse=np.sqrt(mse)
print('RMSE = ',rmse)

mae=mean_absolute_error(y_train,y_train_pred)
print('MAE = ',mae)

score=r2_score(y_train,y_train_pred)
print('Score = ',score*100,'%')
```

# Classification Data :

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
dir(iris)
```

```python
iris.data
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
iris.target
```

```python
df['Species']=iris.target
df
```

# Train_Test_Split

```python
x=df.drop('Species',axis=1)
x
```

```python
x=df.drop('Species',axis=1)
y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=35,test_size=0.3,stratify=y)
```

```python
x_train,y_train
```

```python
ridge_clf=RidgeClassifier()
ridge_clf.fit(x_train,y_train)
```

```python
#Testing Data Evaluation
y_pred=ridge_clf.predict(x_test)
print(y_pred)
cnf_matrix=confusion_matrix(y_test,y_pred)
cnf_matrix

acc_score=accuracy_score(y_test,y_pred)
print('acc_score = ',acc_score)

clf_report=classification_report(y_test,y_pred)
print('clf_report = ',clf_report)
```

```python
#Training Data Evaluation
y_train_pred=ridge_clf.predict(x_train)
print(y_train_pred)
cnf_matrix=confusion_matrix(y_train,y_train_pred)
cnf_matrix

acc_score=accuracy_score(y_train,y_train_pred)
print('acc_score = ',acc_score)

clf_report=classification_report(y_train,y_train_pred)
print('clf_report = ',clf_report)
```

```python
ridge_model=RidgeClassifier()
hyperparameter_grid={'alpha':np.arange(0.05,1.1,0.001)}
gscv_ridge_model=GridSearchCV(ridge_model,hyperparameter_grid,cv=5)
gscv_ridge_model.fit(x_train,y_train)
```
