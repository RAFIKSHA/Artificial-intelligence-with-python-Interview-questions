isna() :
  The isna() method return a DataFrame object where all the values are replaced with a boolean value True for NA
  (Not A Number) values and, otherwise False.
 Syntax :
   dataframe.isna()
     PARAMETER --> This method takes no parameter.
     RETURN VALUE --> Returns A DataFrame with boolean values.

```python
data.isna()
```

```python
data.isna().sum()   
```

```python
data.isnull().sum()
```

dropna() :
  The dropna() method remove the rows that contains null values.
  The dropna() method return a new dataframe object unless the inplace parameter is set to true.

```python
data.dropna()
```

```python
#data=data.dropna()    #Assign Method
data
```

```python
data.dropna(inplace=True)
data
```

Thresold Paramter With Axis() :
  Atleast n columns should have non null value.
          Axis=0--> Along rowwise it iterate column.
          Axis=1--> Along columnwise it iterate row.

```python
data.dropna(thresh=6,axis=0)    #Columns--> Rows Delete
```

```python
data.dropna(thresh=99,axis=1)    #Rows--> Columns Delete
```
