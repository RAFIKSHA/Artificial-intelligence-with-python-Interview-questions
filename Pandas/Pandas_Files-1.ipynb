There are two data structure in pandas :
1.Series (1D-One Dimensional) :
   A series can be seen as one dimensional array.
   The data structure can hold any data type is including string, int, folat      and any python object.
 Syntax :
    pd.Series([1,2,3,4,5])
   Pandas series can also contain a dict.
    d={'a':[10,20],'b':(10,20)}
    pd.Series(d)
2.Data frame : (2D)
   It is 2D type.
   Table with mutable rows and columns.
   Mutable(add,update,delete,rename,replace).
 Application :
  It is used for working on dataset.
  Analysis.

  Manupulation of the dataset.
