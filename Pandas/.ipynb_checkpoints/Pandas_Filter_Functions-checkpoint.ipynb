Unique Function :
  It shows the unique or distinct value/categories in the column.

```python
df['genre'].unique()
```

nunique function :
  It count the number of unique value.

```python
len(df['genre'].unique())
```

```python
df['genre'].nunique()
```

value_counts() :

```python
df['content_rating'].value_counts()
```

isin function() :
  Add required keys in one list and then perform a operation.

```python
a=['NOT RATED','TV-MA']
re_data=df.loc[(df['content_rating'].isin(a))]
re_data
```

```python
re_data.tail(20)
```

```python
Not isin Function() :
```

```python
df.loc[~(df['content_rating'].isin(a))]
```

```python
df.loc[~(df['content_rating'].isin(a))]['content_rating'].unique()
```

```python
df.loc[~(df['content_rating'].isin(a)) & (df['genre']=='Animation')]
#a=['NOT RATED','TV-MA']
```

```python
df.loc[~(df['content_rating'].isin(a)) & (df['genre']=='Animation')]['content_rating'].nunique()
```

```python
df.loc[~(df['content_rating'].isin(a)) & (df['genre']=='Animation')]['content_rating'].unique()
```
