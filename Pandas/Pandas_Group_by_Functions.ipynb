```python
#Aggergate function by using sum,min,max,mean and median in numpy np and mode in pandas pd.Series.
#By using tuple() 
df.groupby(['content_rating'])['duration'].agg((np.mean,np.median,np.sum,np.min,np.max,pd.Series.mode)).reset_index()
```

```python
#By using list[]
df.groupby(['content_rating'])['duration'].agg([np.mean,np.median,np.sum,np.min,np.max,pd.Series.mode]).reset_index()
```

Apply Function :

```python
def upper_case(value):
    value=value.upper()
    return value
```

```python
df['title']=df['title'].apply(upper_case)
df
```

```python
df['title']=df['title'].apply(lambda x:x.upper())
df
```
