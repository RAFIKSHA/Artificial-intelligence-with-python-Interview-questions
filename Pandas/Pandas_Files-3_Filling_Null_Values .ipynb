Filling Null Values :

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Emp_Records.csv')
data
```

Fillna() :
  Filling a single value in the entire data frame.

```python
data.fillna(100)
```

```python
data['Weight in Kgs'].fillna(90)
```

```python
data['Weight in Kgs'].fillna(90,inplace=True)
data
```

```python
mean=np.mean(data['Weight in Kgs'])   #Average value is Weight in Kgs
mean
```

```python
data['Weight in Kgs'].fillna(mean,inplace=True)
data
```

```python
data.fillna(data.mean())
```

```python
median=np.median(data['Weight in Kgs'])   #Average value is Weight in Kgs
median
```

```python
data['Weight in Kgs'].fillna(median,inplace=True)
data
```

```python
data.fillna(data.median())
```
