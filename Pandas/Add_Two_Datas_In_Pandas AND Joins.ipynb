Append :
 When the both the dataframes are same then append it added below the first dataframe.

```python
import pandas as pd
import numpy as np
```

```python
df1=pd.DataFrame({'ASCII NUM':np.arange(65,71),'cha':[chr(i) for i in range(65,71)]})
df1
```

```python
df2=pd.DataFrame({'ASCII NUM':np.arange(69,75),'char':[chr(i) for i in range(69,75)]})
df2
```

```python
#Appending along row
df1.append(df2)
```

Concat :
1.Along Row.
2.Along Column.

```python
display(df1,df2)
```

```python
#1.Along row
pd.concat([df1,df2],axis=0)
```

```python
df3=pd.DataFrame({'ASCII NUM':np.arange(76,82),'cha':[chr(i) for i in range(76,82)]})
df3
```

```python
pd.concat([df1,df3],axis=0)
```

```python
pd.concat([df1,df3],axis=0,ignore_index=True)
```

```python
#2.Along column
pd.concat([df1,df3],axis=1)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df3],axis=1,ignore_index=True)
```

```python
df31=pd.DataFrame({'ASCII NUM':np.arange(76,83),'cha':[chr(i) for i in range(76,83)]})
df31
```

```python
pd.concat([df1,df31],axis=1)
```

Inner Join :
  Commom element.

```python
pd.concat([df1,df31],axis=1,join='inner')
```

Outer Join :
    All element and duplicates are allowed but in the dataframe one element display. 

```python
pd.concat([df1,df31],axis=1,join='outer')
```

Merge :

```python
display(df1,df2)
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='ASCII NUM',how='outer')
```

```python
pd.merge(df1,df2,on='ASCII NUM',how='inner')
```

```python
#Left Join
df1.merge(df2,how='left')
```

```python
df1.merge(df2,on=['ASCII NUM'],how='left')
```

```python
#Right Join
df1.merge(df2,how='right')
```

```python
df1.merge(df2,on=['ASCII NUM'],how='right')
```
